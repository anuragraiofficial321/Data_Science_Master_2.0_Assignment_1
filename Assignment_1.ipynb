### Q3. Explain the use of the following operators using an example:

(i)	    /

(ii)	% 

(iii)	//

(iv)	**

#### Solutions:
 - (i) / This operator is division operator it is used when we want to calcuate division of two numbers
##### Example:

```python
first_number=34
second_number=3
result=first_number/second_number
print("Division of {0} and {1} is {2}".format(first_number,second_number,result))
```

- (2) % This operator is a Modulo operator it is used when we want to get the reminder after deviding the two numbers
##### Example:

```python
first_number=34
second_number=3
result=first_number%second_number
print("Reminder of {0} and {1} is {2}".format(first_number,second_number,result))
```

- (3) // This operator is a floor devision operator it is used when we want to round the result down to the nearest whole number after deviding the two numbers.
##### Example:

```python
first_number=34
second_number=3
result=first_number//second_number
print("Round of {0} and {1} is {2}".format(first_number,second_number,result))
```

- (4) ** This operator is a Exponent operator it is used to perform the exponent arithmetic in python.
##### Example:

```python
first_number=34
second_number=3
result=first_number**second_number
print("Exponential of {0} and {1} is {2}".format(first_number,second_number,result))
```

```python
#To check a number is divisble by other number then we need to use Modulo Operator if Reminder is 0 then that numbers are purely divisible

first_nu=24
second_nu=2
count=0
tempfirst_nu=first_nu
while tempfirst_nu%second_nu==0:
    tempfirst_nu=tempfirst_nu//second_nu
    count+=1
if count==0:
    print("{0} is not divisible by {1}".format(first_nu,second_nu))
else:
    print("{1} is divisible by {2} and can be divided {0} times.".format(count,first_nu,second_nu))


```

### Q6. Create a list containing 25 int type data. Using for loop and if-else condition print if the element is divisible by 3 or not

```python
#creating a list containting 25 int type data using for loop
list_data=[]
for i in range(1,26):
    list_data.append(i)

#checking number is divisible by 3 or not
for i in list_data:
    if i%3==0:   #if reminder is 0 then first number is divisible by 3
        print(f"{i} is divisible by 3")
    else:
        print(f"{i} is not divisible by 3")

```

### Q7. What do you understand about mutable and immutable data types? Give examples for both showing this property

#### Mutable:
- Mutable means after creating a variable with some values and later we can change the values i.e update the values,delete the values and insert the values
- Example: List
#### Immutable:
- Immutable menas after creating a variable with some value and later we can not change the vlaue 
- Example: String,Tuple

#### Example:


```python
# Mutable:
list_value=['Master',1,5.6,True,[4,6,7],(5,6,9),'Data_Science',56,4,'[3,5,6]']
#lets change the Master to Data
list_value[0]="Data"
#add science in last of the list_value
list_value.append("Science")
print(list_value)
```

```python
# Immutable:
list_value=('Master',1,5.6,True,[4,6,7],(5,6,9),'Data_Science',56,4,'[3,5,6]')
#lets change the Master to Data
list_value[0]="Data"    #This will show error
print(list_value)
```

#### Above code is showing error means we can not change the immutable variable values
